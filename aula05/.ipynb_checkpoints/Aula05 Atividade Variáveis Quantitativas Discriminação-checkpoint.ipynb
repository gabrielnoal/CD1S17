## <font color='blue'>Discriminação Salarial</font>

Uma empresa está sendo acusada de pagar um salário maior para os homens do que para as mulheres. 
Para justificar a acusação, apresentou-se uma lista de salários de uma amostra de funcionários. 
Essa empresa tem motivos para se preocupar?

A base de dados `Discriminação.xlsx` apresenta, para cada funcionário, as variáveis:
1. Sexo ('fem' ou 'masc');
2. Anos (anos no emprego); e 
3. Salario (salário mensal, em mil reais). 

Considerando esses dados, faça as seguintes tarefas:

1. Análise descritiva completa considerando as ferramentas estatísticas e computacionais vistas até o momento. Faça essa análise com olhar na preocupação da empresa.
___

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_excel('Discriminação.xlsx')
```

```python
dados.head()
```

```python
dados_masc = dados[dados.Sexo =='masc']
dados_fem = dados[dados.Sexo =='fem']
```

```python
fig = plt.figure(figsize=(10, 5))


plt.subplot(121)
plt.scatter(dados_masc.Anos, dados_masc.Salario, c='blue', alpha=0.8)
plt.title('Anos versus Salário  (Homens)')
plt.xlabel('Anos (anos no emprego)')
plt.ylabel('Salário (salário mensal, em mil reais)')
plt.ylim(20,70)
plt.xlim(-2,30)

plt.subplot(122)
plt.scatter(dados_fem.Anos, dados_fem.Salario, c='pink', alpha=0.8)
plt.title('Anos versus Salário  (Mulheres)')
plt.xlabel('Anos (anos no emprego)')
plt.ylabel('Salário (salário mensal, em mil reais)')
plt.ylim(20,70)
plt.xlim(-2,30)



plt.tight_layout()
plt.show()

print("Covariância entre Anos e Salário (Homens):",np.cov(dados_masc.Anos, dados_masc.Salario)[0,1])
print("Correlação entre Anos e Salário (Homens):",np.corrcoef(dados_masc.Anos, dados_masc.Salario)[0,1])
print()
print("Covariância entre Anos e Salário (Mulheres):",np.cov(dados_fem.Anos, dados_fem.Salario)[0,1])
print("Correlação entre Anos e Salário (Mulheres):",np.corrcoef(dados_fem.Anos, dados_fem.Salario)[0,1])
```

### Analisando a covariância e a correlação dos homens obtemos uma covariância positiva e uma correlação positiva próxima de 0.67.
### Analisando a covariância e a correlação das mulheres obtemos uma covariância positiva e uma correlação positiva próxima de 0.80.

### Indicando que a correlação das mulheres é maior que a dos homens, mas apenas com esses dados não podemos dizer se há ou não descriminação, já que as mulheres ficam menos tempo na empresa e por consequência acabam recebendo menos.


```python
a = dados_masc.Anos.cov(dados_masc.Salario) / dados_masc.Anos.var()
b = dados_masc.Salario.mean() - a*dados_masc.Anos.mean()

c = dados_fem.Anos.cov(dados_fem.Salario) / dados_fem.Anos.var()
d = dados_fem.Salario.mean() - c*dados_fem.Anos.mean()


fig = plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.scatter(dados_masc.Anos, dados_masc.Salario, c='blue', alpha=0.8)
plt.plot((dados_masc.Anos.min(), dados_masc.Anos.max()), (a*dados_masc.Anos.min()+b, a*dados_masc.Anos.max()+b), color='blue')
plt.title('Anos versus Salário  (Mulheres)')
plt.xlabel('Anos (anos no emprego)')
plt.ylabel('Salário (salário mensal, em mil reais)')
plt.ylim(20,70)
plt.xlim(-2,30)


plt.scatter(dados_fem.Anos, dados_fem.Salario, c='red', alpha=0.8)
plt.title('Anos versus Salário  (Mulheres)')
plt.xlabel('Anos (anos no emprego)')
plt.ylabel('Salário (salário mensal, em mil reais)')
plt.ylim(20,70)
plt.xlim(-2,30)
plt.plot((dados_fem.Anos.min(), dados_fem.Anos.max()), (c*dados_fem.Anos.min()+d, c*dados_fem.Anos.max()+d), color='red')


plt.tight_layout()
plt.show()

print('Ajuste linear entre Anos e Salário  (Homens):')
print('Coeficiente linear:', b)
print('Coeficiente angular:', a)
print()
print('Ajuste linear entre Anos e Salário  (Mulheres):')
print('Coeficiente linear:', d)
print('Coeficiente angular:', c)
```

### Para melhor analisar o comportamento do salario das mulheres quando comparados com os anos de trabalho na empresa plotei a disperção de dados das mulheres e dos homens e tracei a reta de ajuste das duas bases de dados no mesmo grafico. Assim podemos concluir que realmente as mulheres trabalham por menos tempo mas se comparadas com os homens que trabalham até o mesmo periodo elas recebem o equivalente.

### Analisando as retas de ajustes vemos que quando começam a trabalhar as é esperado que as mulheres recebam menos que os homens, mas com o passar do tempo os salarios se igualam ou até utrapassam os dos homens. 
## Portanto não há descriminação de genero na empresa.
