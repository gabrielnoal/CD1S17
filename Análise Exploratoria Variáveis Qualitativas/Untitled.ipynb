```python
%matplotlib inline
import math
import matplotlib.pyplot as plt

A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
n=[]

for i in range (1,101):
    n.append(i)
```

```python
for i in range (1,101):
    a = i/((i**2)+1)
    A.append(a)
    #print(a)
print('4.a)A sequência é decrescente e tende a {0}'.format(a))
plt.plot(n,A)
```

```python

for i in range (1,101):
    b = (2*(i**2))/((i**2)+7)
    B.append(b)
#        print(b)
print('4.b)A sequência é crescente e tende a {0}'.format(b))
plt.plot(n,B)
```

```python

for i in range (1,101):
    c = (3*i*math.sqrt(i))/(i*(math.sqrt(i+5)))
    C.append(c)
#        print(c)
print('4.c)A sequência é crescente e tende a {0}'.format(c))

plt.plot(n,C)
```

```python

for i in range (1,101):
    d = (i*math.cos(math.sqrt((i**2)+7)))/((i**2)+1)
    D.append(d)
#        print(a)
print('4.d)A sequência é oscilante e tende a {0}'.format(d))
plt.plot(n,D)
```

```python

for i in range (1,101):
    e = (math.sqrt(i)*(1+8*math.sqrt(i)))/(4*i-1)
    E.append(e)
#        print(e)
print('4.e)A sequência é decrescente e tende a {0}'.format(e)) 
plt.plot(n,E)
```

```python

for i in range (1,101):
    f = ((i**3)-1)/((2*i**3)-i)
    F.append(f)
#    print(f)
print('4.f)A sequência é crescente e tende a {0}'.format(f))
plt.plot(n,F)
```
