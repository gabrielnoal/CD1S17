# Introdução

### No projeto 2, usaremos uma planilha com dados de expectativa de vida ao longo dos anos 1800 e 2013 de vários países, porem filtraremos essa planilha e usaremos apenas 3 desses anos para identificar quais distribuições (funções de densidade de probabilidade - no caso continuo, ou funções de probabilidade - no caso discreto) descrevem melhor cada um desses anos.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats

print('Esperamos trabalhar no diretório')
print(os.getcwd())
dados= pd.read_excel('Life.xlsx')
```

```python
#dados.head()
```

___
# Seleção dos três anos com diferentes formatos de distribuição da expectativa de vida

#### Os anos escolhidos foram: 1919 (fim da Primeira Guerra), 1945 (fim da Segunda Guerra), 1991 (fim da Guerra Fria).
___

```python
dados = dados.dropna()
df = pd.DataFrame()
df['A1919'] = dados.A1919
df['A1945'] = dados.A1945
df['A1991'] = dados.A1991

'''Dataframe Ordenado'''
dfo = pd.DataFrame()
dfo['A1919'] = sorted(dados.A1919)
dfo['A1945'] = sorted(dados.A1945)
dfo['A1991'] = sorted(dados.A1991)
```

___
# Análise Descritiva

```python
print("Describe 1919")
print(df.A1919.describe())
print("Var(1919) = ",df.A1919.var())
print()
print("Describe 1945")
print(df.A1945.describe())
print("Var(1945) = ",df.A1945.var())
print()
print("Describe 1991")
print(df.A1991.describe())
print("Var(1991) = ",df.A1991.var())
```

```python
faixas = np.arange(15,81,3)

fig = plt.figure(figsize=(10,3))

plt.subplot(131)
plt.hist(df.A1919, bins=faixas, normed=1,color="red")
plt.title("Histograma 1919")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.subplot(132)
plt.hist(df.A1945, bins=faixas, normed=1,color="orange")
plt.title("Histograma 1945")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.subplot(133)
plt.hist(df.A1991, bins=faixas, normed=1,color="blue")
plt.title("Histograma 1991")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.tight_layout(2)
plt.show()



fig = plt.figure(figsize=(10,3))

plt.subplot(131)
plt.plot((10,70),(10,70),c="black")
plt.plot(dfo.A1919,dfo.A1945,"ro")
plt.title("QQ Plot 1919x1945")
plt.ylabel("1945")
plt.xlabel("1919")

plt.subplot(132)
plt.plot((10,80),(10,80),c="black")
plt.plot(dfo.A1919,dfo.A1991,"o",c="orange")
plt.title("QQ Plot 1919x1991")
plt.ylabel("1991")
plt.xlabel("1919")

plt.subplot(133)
plt.plot((10,80),(10,80),c="black")
plt.plot(dfo.A1991,dfo.A1945,"bo")
plt.title("QQ Plot 1991x1945")
plt.ylabel("1945")
plt.xlabel("1991")

plt.tight_layout(2)
plt.show()
```

### Conclusão

Com a análise descritiva e os histogramas temos que em 1919 a média era de 33,76 anos com um desvio padrão de 8,2 anos e a expectativa máxima era de 59,98 anos. Em 1945 (26 anos depois) a média subiu para 40,79 com um desvio padrão de 11,23 anos e a expectativa máxima passou a ser 68,46. Então pelo o que parece alguns países se desenvolveram nesse período fazendo com que aumentasse a expectativa média de vida da sua população, mas o desvio padrão também aumentou o que indica que não foram todos os países que aumentaram sua expectativa de vida aumentando assim a desigualdade entra os países que evoluíram pós-guerra e os que não evoluíram. Em 1991 (46 anos depois da segunda guerra) temos números completamente diferentes, a média passou a ser 66,96 anos com desvio padrão de 8,35 anos, o valor máximo é de 79,2 anos e o valor mínimo agora é de é de 46,9 anos, sendo maior que o máximo dos outros dois anos analisados. 

Analisando os histogramas podemos ver como os dois primeiros anos tem assimetria positiva, o que significa que na maioria dos países as pessoas vivem menos e apenas em alguns poucos países as pessoas vivem mais e o último ano tem assimetria negativa demonstrando que naquele ano as pessoas passaram a viver mais que a média mundial.

# Aderência de um modelo probabilístico normal aos dados

```python
faixas = np.arange(15,81,3)

fig = plt.figure(figsize=(10,3))

plt.subplot(131)
mu,sigma = stats.norm.fit(dfo.A1919)
plt.hist(dfo.A1919, bins=faixas, normed=1,color="red")
plt.plot(dfo.A1919,stats.norm.pdf(dfo.A1919,mu,sigma),c="black",lw=2)
plt.title("Normal - 1919")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.subplot(132)
mu,sigma = stats.norm.fit(dfo.A1945)
plt.hist(dfo.A1945, bins=faixas, normed=1,color="orange")
plt.plot(dfo.A1945,stats.norm.pdf(dfo.A1945,mu,sigma),c="black",lw=2)
plt.title("Normal - 1945")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.subplot(133)
mu,sigma = stats.norm.fit(dfo.A1991)
plt.hist(dfo.A1991, bins=faixas, normed=1,color="blue")
plt.plot(dfo.A1991,stats.norm.pdf(dfo.A1991,loc=mu,scale=sigma),color="black",lw=2)
plt.title("Normal - 1991")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.tight_layout(2)
plt.show()
```

```python

fra1919 = []
for i in range(len(dfo.A1919)):
    fra1919.append((i+0.5)/len(dfo.A1919))

fra1945 = []
for i in range(len(dfo.A1945)):
    fra1945.append((i+0.5)/len(dfo.A1945))

fra1991 = []
for i in range(len(dfo.A1991)):
    fra1991.append((i+0.5)/len(dfo.A1991))

    
q1919 = []
q1945 = []
q1991 = []

mu,sigma = stats.norm.fit(dfo.A1919)
for p in fra1919:
    q1919.append(stats.norm.ppf(p,mu,sigma))

mu,sigma = stats.norm.fit(dfo.A1945)
for p in fra1945:
    q1945.append(stats.norm.ppf(p,mu,sigma))

mu,sigma = stats.norm.fit(dfo.A1991)
for p in fra1991:    
    q1991.append(stats.norm.ppf(p,mu,sigma))


fig = plt.figure(figsize=(12,4))

plt.subplot(131)
plt.plot(dfo.A1919, q1919, 'o',c="red", lw=3)
plt.plot((10,60),(10,60), 'black',alpha=0.8)
plt.title('1919')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')

plt.subplot(132)
plt.plot(dfo.A1945, q1945, 'o',c="orange", lw=3)
plt.plot((0,90),(0,90), 'black', alpha=0.8)
plt.title('1945')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')

plt.subplot(133)
plt.plot(dfo.A1991, q1991, 'o',c="blue", lw=3)
plt.plot((40,100),(40,100), 'black', alpha=0.8)
plt.title('1991')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')


plt.tight_layout(2)
plt.show()
```

```python
cdf1919 = []
cdf1945 = []
cdf1991 = []

mu,sigma = stats.norm.fit(dfo.A1919)
for x in q1919:
    cdf1919.append(stats.norm.cdf(x,mu,sigma))

    
mu,sigma = stats.norm.fit(dfo.A1945)
for x in q1945:
    cdf1945.append(stats.norm.cdf(x,mu,sigma))

    
mu,sigma = stats.norm.fit(dfo.A1991)
for x in q1991:
    cdf1991.append(stats.norm.cdf(x,mu,sigma))

fig = plt.figure(figsize=(13,4))

plt.subplot(131)
mu,sigma = stats.norm.fit(dfo.A1919)
plt.plot(dfo.A1919,stats.norm.cdf(dfo.A1919,mu,sigma),c="black",lw=5)
plt.plot(dfo.A1919,cdf1919, 'o',c="red", alpha=0.4,markersize=10)
plt.title("1919")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")

plt.subplot(132)
mu,sigma = stats.norm.fit(dfo.A1945)
plt.plot(dfo.A1945,stats.norm.cdf(dfo.A1945,mu,sigma),c="black",lw=5)
plt.plot(dfo.A1945,cdf1945, 'o',c="orange", alpha=0.4,markersize=10)
plt.title("1945")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")

plt.subplot(133)
mu,sigma = stats.norm.fit(dfo.A1991)
plt.plot(dfo.A1991,stats.norm.cdf(dfo.A1991,mu,sigma),c="black",lw=5)
plt.plot(dfo.A1991,cdf1991, 'o',c="blue", alpha=0.4,markersize=10)
plt.title("1991")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")


plt.tight_layout()
plt.show()
```

# Aderência de um modelo probabilístico não normal aos dados

```python
faixas = np.arange(15,81,3)

fig = plt.figure(figsize=(10,3))

plt.subplot(131)
s,l,e = stats.lognorm.fit(dfo.A1919)
plt.hist(dfo.A1919, bins=faixas, normed=1,color="red")
plt.plot(dfo.A1919,stats.lognorm.pdf(dfo.A1919,s,l,e),c="black",lw=2)
plt.title("Log normal - 1919")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.subplot(132)
s,l,e = stats.lognorm.fit(dfo.A1945)
plt.hist(dfo.A1945, bins=faixas, normed=1,color="orange")
plt.plot(dfo.A1945,stats.lognorm.pdf(dfo.A1945,s,l,e),c="black",lw=2)
plt.title("Log normal - 1945")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.subplot(133)
a,b,l,e = stats.beta.fit(dfo.A1991)
plt.hist(dfo.A1991, bins=faixas, normed=1,color="blue")
plt.plot(dfo.A1991,stats.beta.pdf(dfo.A1991,a,b,l,e),color="black",lw=2)
plt.title("Beta - 1991")
plt.xlabel("Expectativa de vida ( em anos)")
plt.ylabel("Densidade")
plt.ylim(0,0.12)

plt.tight_layout(2)
plt.show()
```

```python

fra1919 = []
for i in range(len(dfo.A1919)):
    fra1919.append((i+0.5)/len(dfo.A1919))

fra1945 = []
for i in range(len(dfo.A1945)):
    fra1945.append((i+0.5)/len(dfo.A1945))

fra1991 = []
for i in range(len(dfo.A1991)):
    fra1991.append((i+0.5)/len(dfo.A1991))

    
q1919 = []
q1945 = []
q1991 = []

s,l,e = stats.lognorm.fit(dfo.A1919)
for p in fra1919:
    q1919.append(stats.lognorm.ppf(p,s,l,e))

s,l,e = stats.lognorm.fit(dfo.A1945)
for p in fra1945:
    q1945.append(stats.lognorm.ppf(p, s, loc=l, scale=e))

a,b,l,e = stats.beta.fit(dfo.A1991)
for p in fra1991:    
    q1991.append(stats.beta.ppf(p, a,b,l,e))


fig = plt.figure(figsize=(12,4))

plt.subplot(131)
plt.plot(dfo.A1919, q1919, 'o',c="red", lw=3)
plt.plot((10,70),(10,70), 'black',alpha=0.8)
plt.title('1919')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')

plt.subplot(132)
plt.plot(dfo.A1945, q1945, 'o',c="orange", lw=3)
plt.plot((10,90),(10,90), 'black', alpha=0.8)
plt.title('1945')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')

plt.subplot(133)
plt.plot(dfo.A1991, q1991, 'o',c="b", lw=3)
plt.plot((40,90),(40,90), 'black', alpha=0.8)
plt.title('1991')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')


plt.tight_layout(2)
plt.show()
```

```python
cdf1919 = []
cdf1945 = []
cdf1991 = []

s,l,e = stats.lognorm.fit(dfo.A1919)
for x in q1919:
    cdf1919.append(stats.lognorm.cdf(x,s,l,e))

    
s,l,e = stats.lognorm.fit(dfo.A1945)
for x in q1945:
    cdf1945.append(stats.lognorm.cdf(x,s,l,e))

    
a,b,l,e = stats.beta.fit(dfo.A1991)
for x in q1991:
    cdf1991.append(stats.beta.cdf(x,a,b,l,e))
    

    
fig = plt.figure(figsize=(13,4))

plt.subplot(131)
s,l,e = stats.lognorm.fit(dfo.A1919)
plt.plot(dfo.A1919,stats.lognorm.cdf(dfo.A1919,s,l,e),color="black",lw=5)
plt.plot(dfo.A1919,cdf1919, 'go',color="red",markersize=10, alpha=0.4)
plt.title("1919")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")

plt.subplot(132)
s,l,e = stats.lognorm.fit(dfo.A1945)
plt.plot(dfo.A1945,stats.lognorm.cdf(dfo.A1945,s,l,e),color="black",lw=5)
plt.plot(dfo.A1945,cdf1945, 'go',color="orange",markersize=10, alpha=0.4)
plt.title("1945")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")

plt.subplot(133)
a,b,l,e = stats.beta.fit(dfo.A1991)
plt.plot(dfo.A1991,stats.beta.cdf(dfo.A1991,a,b,l,e),color="black",lw=5)
plt.plot(dfo.A1991,cdf1991, 'go',color="blue",markersize=10, alpha=0.4)
plt.title("1991")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")


plt.tight_layout()
plt.show()
```

### Conclusão

Na figura acima temos 15 gráficos, nos três primeiros comparo qual distribuição melhor se adequa aos dados. No primeiro e no segundo (1919 e 1945) usei a distribuição normal e a log-normal, como pedido no enunciado do projeto. No terceiro (1991) usei a normal, mas ela não me parecia adequada aos dados, então decidi usar a distribuição beta.

Para ter certeza qual das distribuições melhor se adequava fiz duas comparações para cada ano, a dos quantis teóricos versus observados (QQ-Plot) e a da probabilidade acumulada.

### Cálculo de probabilidades a partir da distribuição normal

#### Probabilidade de um país ter expectativa de vida superior a 70 anos.

```python
a =round(((1 - stats.norm.cdf(70, loc=df.A1919.mean(), scale=df.A1919.std()))*100),2)
b =round(((1 - stats.norm.cdf(70, loc=df.A1945.mean(), scale=df.A1945.std()))*100),2)
c =round(((1 - stats.norm.cdf(70, loc=df.A1991.mean(), scale=df.A1991.std()))*100),2)
print("No ano de 1919 a probabilidade de um país ter expectativa de vida superior a 70 anos é de {0}%".format(a))
print("No ano de 1945 a probabilidade de um país ter expectativa de vida superior a 70 anos é de {0}%".format(b))
print("No ano de 1991 a probabilidade de um país ter expectativa de vida superior a 70 anos é de {0}%".format(c))
```

#### Sabendo que um país tem expectiva de vida superior a 70 anos, qual a probabilidade desse possuir expectativa inferior a 75 anos?

```python
a =round(((stats.norm.cdf(75, loc=df.A1919.mean(), scale=df.A1919.std())-stats.norm.cdf(70, loc=df.A1919.mean(), scale=df.A1919.std()))/(stats.norm.cdf(80, loc=df.A1919.mean(), scale=df.A1919.std())-stats.norm.cdf(70, loc=df.A1919.mean(), scale=df.A1919.std()))*100),2)
b =round(((stats.norm.cdf(75, loc=df.A1945.mean(), scale=df.A1945.std())-stats.norm.cdf(70, loc=df.A1945.mean(), scale=df.A1945.std()))/(stats.norm.cdf(80, loc=df.A1945.mean(), scale=df.A1945.std())-stats.norm.cdf(70, loc=df.A1945.mean(), scale=df.A1945.std()))*100),2)
c =round(((stats.norm.cdf(75, loc=df.A1991.mean(), scale=df.A1991.std())-stats.norm.cdf(70, loc=df.A1991.mean(), scale=df.A1991.std()))/(stats.norm.cdf(80, loc=df.A1991.mean(), scale=df.A1991.std())-stats.norm.cdf(70, loc=df.A1991.mean(), scale=df.A1991.std()))*100),2)
print("Sabendo que um país no ano de 1919 tem expectiva de vida superior a 70 anos, qual a probabilidade desse possuir expectativa inferior a 75 anos é de {0}%".format(a))
print("Sabendo que um país no ano de 1919 tem expectiva de vida superior a 70 anos, qual a probabilidade desse possuir expectativa inferior a 75 anos é de {0}%".format(b))
print("Sabendo que um país no ano de 1919 tem expectiva de vida superior a 70 anos, qual a probabilidade desse possuir expectativa inferior a 75 anos é de {0}%".format(c))
```
