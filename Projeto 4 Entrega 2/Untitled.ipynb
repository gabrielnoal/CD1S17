### Descrição do tema:

    Para este projeto, iremos analisar os acidentes de transito e verificar qual genero comete mais acidentes nas estradas.
    Para isso usaremos um base de dados aberta de acidentes coletadas pela Policia Rodviaria Federal de todo o pais.


```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats
import math
import matplotlib.mlab as mlab

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados2015 = pd.read_excel('acidentes_2015.xlsx')
dados2016 = pd.read_excel('acidentes_2016.xlsx')
```

```python
dados2015_masc = dados2015[(dados2015.idade >= 17) & (dados2015.sexo == "Masculino") & (dados2015.tipo_envolvido == "Condutor") & (dados2015.tipo_veiculo == "Automóvel")]
dados2015_fem = dados2015[(dados2015.idade >= 17) & (dados2015.sexo == "Feminino") & (dados2015.tipo_envolvido == "Condutor") & (dados2015.tipo_veiculo == "Automóvel")]

dados2016_masc = dados2016[(dados2016.idade >= 17) & (dados2016.sexo == "M") & (dados2016.tipo_envolvido == "Condutor") & (dados2016.tipo_veiculo == "Automóvel")]
dados2016_fem = dados2016[(dados2016.idade >= 17) & (dados2016.sexo == "F") & (dados2016.tipo_envolvido == "Condutor") & (dados2016.tipo_veiculo == "Automóvel")]

```

```python
df_2015_masc_morto = dados2015_masc[(dados2015_masc.estado_fisico == "Morto       ")]
df_2015_fem_morto = dados2015_fem[(dados2015_fem.estado_fisico == "Morto       ")]

df_2016_masc_morto = dados2016_masc[(dados2016_masc.estado_fisico == "Morto")]
df_2016_fem_morto = dados2016_fem[(dados2016_fem.estado_fisico == "Morto")]

df_2015_masc_grav = dados2015_masc[(dados2015_masc.estado_fisico == "Ferido Grave")]
df_2015_fem_grav = dados2015_fem[(dados2015_fem.estado_fisico == "Ferido Grave")]

df_2016_masc_grav = dados2016_masc[(dados2016_masc.estado_fisico == "Ferido Grave")]
df_2016_fem_grav = dados2016_fem[(dados2016_fem.estado_fisico == "Ferido Grave")]

df_2015_masc_lev = dados2015_masc[(dados2015_masc.estado_fisico == "Ferido Leve ")]
df_2015_fem_lev = dados2015_fem[(dados2015_fem.estado_fisico == "Ferido Leve ")]

df_2016_masc_lev = dados2016_masc[(dados2016_masc.estado_fisico == "Ferido Leve")]
df_2016_fem_lev = dados2016_fem[(dados2016_fem.estado_fisico == "Ferido Leve")]

df_2015_masc_ile = dados2015_masc[(dados2015_masc.estado_fisico == "Ileso       ")]
df_2015_fem_ile = dados2015_fem[(dados2015_fem.estado_fisico == "Ileso       ")]

df_2016_masc_ile = dados2016_masc[(dados2016_masc.estado_fisico == "Ileso")]
df_2016_fem_ile = dados2016_fem[(dados2016_fem.estado_fisico == "Ileso")]

```

```python
fra2015_h_morto = []
q2015_h_morto = []
for i in range(len(dfo_2015_masc_morto)):
    fra2015_h_morto.append((i+0.5)/len(dfo_2015_masc_morto))
    
mu,sigma = stats.norm.fit(dfo_2015_masc_morto)
for p in fra2015_h_morto:
    q2015_h_morto.append(stats.norm.ppf(p,mu,sigma))
    
plt.plot(dfo_2015_masc_morto, q2015_h_morto, 'o',c="red", lw=3)
plt.plot((-20,100),(-20,100), 'black',alpha=0.8)
```
