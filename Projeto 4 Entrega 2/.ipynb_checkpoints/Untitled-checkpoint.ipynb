### Descrição do tema:

    Para este projeto, iremos analisar os acidentes de transito e verificar qual genero comete mais acidentes nas estradas.
    Para isso usaremos um base de dados aberta de acidentes coletadas pela Policia Rodviaria Federal de todo o pais.


```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats
import math
import matplotlib.mlab as mlab

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados2015 = pd.read_excel('acidentes_2015.xlsx')
dados2016 = pd.read_excel('acidentes_2016.xlsx')
```

```python
dados2015_masc = dados2015[(dados2015.idade >= 17) & (dados2015.sexo == "Masculino") & (dados2015.tipo_envolvido == "Condutor") & (dados2015.tipo_veiculo == "Automóvel")]
dados2015_fem = dados2015[(dados2015.idade >= 17) & (dados2015.sexo == "Feminino") & (dados2015.tipo_envolvido == "Condutor") & (dados2015.tipo_veiculo == "Automóvel")]

dados2016_masc = dados2016[(dados2016.idade >= 17) & (dados2016.sexo == "M") & (dados2016.tipo_envolvido == "Condutor") & (dados2016.tipo_veiculo == "Automóvel")]
dados2016_fem = dados2016[(dados2016.idade >= 17) & (dados2016.sexo == "F") & (dados2016.tipo_envolvido == "Condutor") & (dados2016.tipo_veiculo == "Automóvel")]

```

```python
df_2015_masc_morto = dados2015_masc[(dados2015_masc.estado_fisico == "Morto       ")]
df_2015_fem_morto = dados2015_fem[(dados2015_fem.estado_fisico == "Morto       ")]

df_2016_masc_morto = dados2016_masc[(dados2016_masc.estado_fisico == "Morto")]
df_2016_fem_morto = dados2016_fem[(dados2016_fem.estado_fisico == "Morto")]

df_2015_masc_grav = dados2015_masc[(dados2015_masc.estado_fisico == "Ferido Grave")]
df_2015_fem_grav = dados2015_fem[(dados2015_fem.estado_fisico == "Ferido Grave")]

df_2016_masc_grav = dados2016_masc[(dados2016_masc.estado_fisico == "Ferido Grave")]
df_2016_fem_grav = dados2016_fem[(dados2016_fem.estado_fisico == "Ferido Grave")]

df_2015_masc_lev = dados2015_masc[(dados2015_masc.estado_fisico == "Ferido Leve ")]
df_2015_fem_lev = dados2015_fem[(dados2015_fem.estado_fisico == "Ferido Leve ")]

df_2016_masc_lev = dados2016_masc[(dados2016_masc.estado_fisico == "Ferido Leve")]
df_2016_fem_lev = dados2016_fem[(dados2016_fem.estado_fisico == "Ferido Leve")]

df_2015_masc_ile = dados2015_masc[(dados2015_masc.estado_fisico == "Ileso       ")]
df_2015_fem_ile = dados2015_fem[(dados2015_fem.estado_fisico == "Ileso       ")]

df_2016_masc_ile = dados2016_masc[(dados2016_masc.estado_fisico == "Ileso")]
df_2016_fem_ile = dados2016_fem[(dados2016_fem.estado_fisico == "Ileso")]
```

```python
faixas= range(15,90,5)

fig = plt.figure(figsize=(10,5))

plt.subplot(221)
df_2015_masc_morto.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Mortos 2015")


plt.subplot(222)
df_2015_fem_morto.idade.hist(bins=faixas,normed=1, color='red')
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Mortas 2015")

plt.subplot(223)
df_2016_masc_morto.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Mortos 2016")

plt.subplot(224)
df_2016_fem_morto.idade.hist(bins=faixas,normed=1, color='red')
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Mortas 2016")


plt.tight_layout()
plt.show()


'''Gravemente Ferido'''


fig = plt.figure(figsize=(10,5))

plt.subplot(221)
df_2015_masc_grav.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Gravementes Feridos 2015")


plt.subplot(222)
df_2015_fem_grav.idade.hist(bins=faixas,normed=1, color='red')
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Gravementes Feridas 2015")

plt.subplot(223)
df_2016_masc_grav.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Gravementes Feridos 2016")

plt.subplot(224)
df_2016_fem_grav.idade.hist(bins=faixas,normed=1, color='red')
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Gravementes Feridas 2016")


plt.tight_layout()
plt.show()

'''Levemente Ferido'''


fig = plt.figure(figsize=(10,5))

plt.subplot(221)
df_2015_masc_lev.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Levemente Feridos 2015")


plt.subplot(222)
df_2015_fem_lev.idade.hist(bins=faixas,normed=1, color='red')
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Levemente Feridas 2015")

plt.subplot(223)
df_2016_masc_lev.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Levemente Feridos 2016")

plt.subplot(224)
df_2016_fem_lev.idade.hist(bins=faixas,normed=1, color='red')
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Levemente Feridas 2016")


plt.tight_layout()
plt.show()

'''Ileso'''


fig = plt.figure(figsize=(10,5))

plt.subplot(221)
df_2015_masc_ile.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Ilesos 2015")


plt.subplot(222)
df_2015_fem_ile.idade.hist(bins=faixas,normed=1, color='red')
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Ilesas 2015")

plt.subplot(223)
df_2016_masc_ile.idade.hist(bins=faixas,normed=1)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Homens Ilesos 2016")

plt.subplot(224)
df_2016_fem_ile.idade.hist(bins=faixas,normed=1, color='red',alpha=0.6)
plt.ylim(0,0.045)
plt.xlabel("Idades")
plt.ylabel("Densidade")
plt.title("Histograma Mulheres Ilesas 2016")




plt.tight_layout()
plt.show()
```
