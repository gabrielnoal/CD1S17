### Descrição do tema:

    Para este projeto, iremos analisar os acidentes de transito e verificar se a idade média dos homens que causam acidentes de carro e morrem é a mesma que a das mulheres na mesma situação nos anos 2015 e 2016.
    Para isso usaremos uma base de dados aberta de acidentes coletadas pela Polícia Rodoviária Federal de todo o pais.
   
### Analiticamente:

##### Hipótese nula: A média das idades dos homens mortos em acidentes de carro é igual a das mulheres.
##### Hipótese alternativa: A média das idades dos homens mortos em acidentes de carro não é igual a das mulheres.



### Estatisticamente:
##### Ho: μ(H) = μ(M)
##### Ha: μ(H) ≠ μ(M)

##### Com um nivel de significancia de 5%

___


```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats
import math
import matplotlib.mlab as mlab
import random

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados2015 = pd.read_excel('acidentes_2015.xlsx')
dados2016 = pd.read_excel('acidentes_2016.xlsx')
```

```python
dados2015_masc = dados2015[(dados2015.idade >= 17) & (dados2015.sexo == "Masculino") & (dados2015.tipo_envolvido == "Condutor") & (dados2015.tipo_veiculo == "Automóvel") & (dados2015.estado_fisico == "Morto       ")]
dados2015_fem = dados2015[(dados2015.idade >= 17) & (dados2015.sexo == "Feminino") & (dados2015.tipo_envolvido == "Condutor") & (dados2015.tipo_veiculo == "Automóvel") & (dados2015.estado_fisico == "Morto       ")]

dados2016_masc = dados2016[(dados2016.idade >= 17) & (dados2016.sexo == "M") & (dados2016.tipo_envolvido == "Condutor") & (dados2016.tipo_veiculo == "Automóvel") & (dados2016.estado_fisico == "Morto")]
dados2016_fem = dados2016[(dados2016.idade >= 17) & (dados2016.sexo == "F") & (dados2016.tipo_envolvido == "Condutor") & (dados2016.tipo_veiculo == "Automóvel") & (dados2016.estado_fisico == "Morto")]

```

```python
H2015 = pd.DataFrame()
H2015['idade'] = dados2015_masc.idade

H2016 = pd.DataFrame()
H2016['idade'] = dados2016_masc.idade

M2015 = pd.DataFrame()
M2015['idade'] = dados2015_fem.idade

M2016 = pd.DataFrame()
M2016['idade'] = dados2016_fem.idade
```

```python
H2015_ord = pd.DataFrame()
H2015_ord['idade'] = sorted(dados2015_masc.idade)

H2016_ord = pd.DataFrame()
H2016_ord['idade'] = sorted(dados2016_masc.idade)

M2015_ord = pd.DataFrame()
M2015_ord['idade'] = sorted(dados2015_fem.idade)

M2016_ord = pd.DataFrame()
M2016_ord['idade'] = sorted(dados2016_fem.idade)
```

```python
nH2015 = len(H2015)
nM2015 = len(M2015)

nH2016 = len(H2016)
nM2016 = len(M2016)

listaH2015 = sorted(H2015.idade)
listaM2015 = sorted(M2015.idade)
listaH2016 = sorted(H2016.idade)
listaM2016 = sorted(M2016.idade)

ListaTotal2015 = listaH2015 + listaM2015
ListaTotal2016 = listaH2016 + listaM2016


p = 10000

dif_perm2015 = []
for i in range (0,p):
    random.shuffle(ListaTotal2015)
    a = np.mean(ListaTotal2015[0:(len(H2015)-1)])
    #print(a)
    b = np.mean(ListaTotal2015[(len(H2015)-1):])
    #print(b)
    dif_perm2015.append(a-b)

    
dif_perm2016 = []
for i in range (0,p):
    random.shuffle(ListaTotal2016)
    a = np.mean(ListaTotal2016[0:(len(H2016)-1)])
    #print(a)
    b = np.mean(ListaTotal2016[(len(H2016)-1):])
    #print(b)
    dif_perm2016.append(a-b)
```

```python
    
print("Média da dif_perm2015",np.mean(dif_perm2015))
print("Média da dif_perm2016",np.mean(dif_perm2016))
```

```python
fig = plt.figure(figsize=(10,3))

plt.subplot(121)
difo_perm2015 = sorted(dif_perm2015)
plt.hist(dif_perm2015, bins=40, normed=1)
plt.plot(difo_perm2015, stats.norm.pdf(difo_perm2015, loc=np.mean(dif_perm2015), scale=np.std(dif_perm2015)), lw = 2, color="black")
plt.title("Histograma Permutado 2015")
plt.xlabel("Diferença das médias (em anos)")
plt.ylabel("Densidade")

plt.subplot(122)
difo_perm2016 = sorted(dif_perm2016)
plt.hist(dif_perm2016, bins=40, normed=1)
plt.plot(difo_perm2016, stats.norm.pdf(difo_perm2016, loc=np.mean(dif_perm2016), scale=np.std(dif_perm2016)), lw = 2, color="black")
plt.title("Histograma Permutado 2016")
plt.xlabel("Diferença das médias (em anos)")
plt.ylabel("Densidade")

plt.tight_layout(1)
plt.show()
```

```python
alpha = 5/100
dif_real2015 = H2015.idade.mean() - M2015.idade.mean()
dif_real2016 = H2016.idade.mean() - M2016.idade.mean()

xcritico2015 = round(((stats.norm.ppf((1-alpha), loc=np.mean(dif_perm2015), scale=np.std(dif_perm2015)))),2)
xcritico2016 = round(((stats.norm.ppf((1-alpha), loc=np.mean(dif_perm2016), scale=np.std(dif_perm2016)))),2)

print("Como em 2015 Xcritico é igual a {0} anos e a diferencia real entre as médias é igual a {1}, rejeitamos Ho. O que significa que a média das idades de homens e mulheres que cauzam acidentes e morrem não são iguais".format(xcritico2015,dif_real2015))
print()
print("Como em 2016 Xcritico é igual a {0} anos e a diferencia real entre as médias é igual a {1}, rejeitamos Ho. O que significa que a média das idades de homens e mulheres que cauzam acidentes e morrem não são iguais".format(xcritico2016,dif_real2016))
```
