```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np

print('Esperamos trabalhar no diretório')
print(os.getcwd())
dados= pd.read_excel('Padaria.xlsx')
```

```python
dados.head()
```

```python
dt = pd.DataFrame()
dt['A'] = sorted(dados.A)
dt['B'] = sorted(dados.B)
```

# 1)

```python
print('Describe Renda A')
print(dt.A.describe())
print('Var(A): ',dt.A.var())
print()
print('Describe Renda B')
print(dt.B.describe())
print('Var(B): ',dt.B.var())
print()
```

# 2)

```python
faixas = np.arange(9,20,1)

fig = plt.figure(figsize=(10,3))

plt.subplot(131)
plt.hist(dt.A, bins=faixas, normed=1,color="red")
plt.title("Histograma Renda A")

plt.subplot(132)
plt.hist(dt.B, bins=faixas, normed=1)
plt.title("Histograma Renda B")
plt.ylim(0,0.6)

plt.subplot(133)
plt.hist(dt.B, bins=faixas, normed=1,color='blue')
plt.hist(dt.A, bins=faixas, normed=1,color="red",alpha=0.7)
plt.title("Histograma Renda A e B")
plt.ylim(0,0.6)

plt.tight_layout()
plt.show()
```

## Conclusão:
O melhor bairro para se construir a padaria de acordo com os histogramas seria o bairro A pois a dispersão de renda é menor que a do bairro B e a média do bairro A é maior que a do bairro B.


# 3)

```python
plt.plot((6,20),(6,20))
plt.plot(dt.A,dt.B,"*",markersize=10)
plt.title("QQ Plot AxB")
plt.xlabel("Renda B")
plt.ylabel("Renda A")
```

## Conclusão:
Analisando o QQ Plot vemos que até o terceiro quartil os valores da renda do bairro A é maior que do bairro B, porem depois do terceiro quartil os valores da renda do bairro B são maiores, indicando que o bairro B é mais rico. Portanto deveríamos construir a padaria no bairro B.


# 4.a)

```python
fig = plt.figure(figsize=(12,4))

plt.subplot(131)
plt.plot(dt.A, stats.norm.pdf(dt.A, loc=dt.A.mean(), scale=dt.A.std()),c='black')
plt.hist(dt.A, normed=1,color="red",alpha=0.5)
plt.title('Normal A')
plt.ylabel("Densidade")
plt.xlabel("Renda familiar (em mil reais)")

plt.subplot(132)
plt.plot(dt.B, stats.norm.pdf(dt.B, loc=dt.B.mean(), scale=dt.B.std()),c='black')
plt.hist(dt.B, normed=1,color="blue",alpha=0.5)
plt.ylabel("Densidade")
plt.xlabel("Renda familiar (em mil reais)")
plt.title('Normal B')

plt.subplot(133)
plt.plot(dt.A, stats.norm.pdf(dt.A, loc=dt.A.mean(), scale=dt.A.std()),c='black')
plt.hist(dt.A, normed=1,color="red",alpha=0.5)
plt.plot(dt.B, stats.norm.pdf(dt.B, loc=dt.B.mean(), scale=dt.B.std()),c='black')
plt.hist(dt.B, normed=1,color="blue",alpha=0.5)
plt.title('Normal A e B')
plt.ylabel("Densidade")
plt.xlabel("Renda familiar (em mil reais)")
plt.xlim(8,20)
plt.ylim(0,0.7)

plt.tight_layout()
plt.show()
```

# 4.b)

```python
fra = []

for i in range(len(dados)):
    fra.append((i+0.5)/len(dados))
    
qA = []
qB = []

for p in fra:
    qA.append(stats.norm.ppf(p, loc=dados.A.mean(), scale=dados.A.std()))
    qB.append(stats.norm.ppf(p, loc=dados.B.mean(), scale=dados.B.std()))
    
fig = plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(dt.A, qA, 'go', lw=3)
plt.plot((13,18),(13,18), 'b-',alpha=0.8)
plt.title('Renda A')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')

plt.subplot(122)
plt.plot(dt.B, qB, 'go', lw=3)
plt.plot((8,20),(8,20), 'b-', alpha=0.8)
plt.title('Renda B')
plt.xlabel('Quantil observado')
plt.ylabel('Quantil teórico')

plt.tight_layout(2)
plt.show()
```

# 4.c)

```python
cdfA = []
cdfB = []

for x in qA:
    cdfA.append(stats.norm.cdf(x, loc=dt.A.mean(), scale=dt.A.std()))

for x in qB:
    cdfB.append(stats.norm.cdf(x, loc=dt.B.mean(), scale=dt.B.std()))
    

fig = plt.figure(figsize=(13,4))

plt.subplot(121)
plt.plot(dt.A,stats.norm.cdf(dt.A, loc=dt.A.mean(), scale=dt.A.std()),lw=5)
plt.plot(dt.A,cdfA, 'go', alpha=0.4)
plt.title("Renda A")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")

plt.subplot(122)
plt.plot(dt.B,stats.norm.cdf(dt.B, loc=dt.B.mean(), scale=dt.B.std()),lw=5)
plt.plot(dt.B,cdfB, 'go', alpha=0.4)
plt.title("Renda B")
plt.ylabel("Prob. Acumulada")
plt.xlabel("Renda familiar (em mil reais)")


plt.tight_layout()
plt.show()
```

# 5.A)

```python
a =round(((1 - stats.norm.cdf(17, loc=dt.A.mean(), scale=dt.A.std()))*100),2)
b=round(((1 - stats.norm.cdf(17, loc=dt.B.mean(), scale=dt.B.std()))*100),2)
print("A probabilidade de uma familia do bairro A ter mais de 17mil reais é de {0}%".format(a))
print("A probabilidade de uma familia do bairro B ter mais de 17mil reais é de {0}%".format(b))
```

# 5.B)

```python
a =round(((stats.norm.cdf(16, loc=dados.A.mean(), scale=dados.A.std())-stats.norm.cdf(15, loc=dados.A.mean(), scale=dados.A.std()))/(stats.norm.cdf(18, loc=dados.A.mean(), scale=dados.A.std())-stats.norm.cdf(15, loc=dados.A.mean(), scale=dados.A.std()))*100),2)
b =round(((stats.norm.cdf(16, loc=dados.B.mean(), scale=dados.B.std())-stats.norm.cdf(15, loc=dados.B.mean(), scale=dados.B.std()))/(stats.norm.cdf(18, loc=dados.B.mean(), scale=dados.B.std())-stats.norm.cdf(15, loc=dados.B.mean(), scale=dados.B.std()))*100),2)
print("Considerando apenas as familias com renda maior que 15mil, a probabilidade de ser menor que 16 mil no bairro A é de {0}%".format(a))
print("Considerando apenas as familias com renda maior que 15mil, a probabilidade de ser menor que 16 mil no bairro B é de {0}%".format(b))
```

# 5.C)

```python
a =((stats.norm.ppf(0.95, loc=dt.A.mean(), scale=dt.A.std())))
b =((stats.norm.ppf(0.95, loc=dt.B.mean(), scale=dt.B.std())))
print("A renda mais baixa dos 5% mais ricos do bairro A é de {0}".format(a))
print("A renda mais baixa dos 5% mais ricos do bairro B é de {0}".format(b))
```
