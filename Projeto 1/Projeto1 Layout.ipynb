## Conclusão 

Pensei em analisar quem são os domicílios que ainda utiliza internet discada e compara-lo com os que usam internet móvel e internet por banda, além disso comparei como os dados se alteraram entre 2014 e 2015.

Primeiro comparei zona rural com zona urbana e constatei que a internet discada aumentou cerca de 3,8% na zona rural. Já a móvel e a banda também tiveram um aumento, porém não tão significativo se comparado com o da discada.

Despois optei por analisar por regiões do pais e ver em qual delas mais se encontra domicílios que utilizam de internet discada. Percebi que de todas as regiões a que tem mais gente que utiliza internet discada é a região sudeste, esta é também a única região que apresenta um aumento no número de usuários de internet discada.

Porém não vamos focar apenas na região sudeste, continuaremos a analisar todas as regiões. Filtrando apenas a zona rural de todas as regiões, podemos ver que a região sul teve um aumento de 54,2% enquanto no mesmo período de tempo os usuários de rede móvel e discada caíram.
Filtrando outra vez, mas agora apenas os usuários da zona urbana de cada região vemos que de todas elas o Sudeste foi a que teve o maior aumento de cerca de 6,4%.
Decidi também estudar a renda per capta dos domicílios para ver se encontrava algo relevante.
É normal quando fazemos um histograma de renda que ele tenha uma assimetria positiva, já que muita gente ganha pouco e pouca gente ganha muito. Porém não consegui concluir nada analisando os histogramas.
